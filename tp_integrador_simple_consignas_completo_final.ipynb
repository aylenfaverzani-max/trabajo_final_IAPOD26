## Datos del trabajo

Vamos a usar una lista de diccionarios.  
Cada diccionario representa a un estudiante.

```python
estudiantes = [
    {"nombre": "Ana", "edad": 22, "trabaja": "Sí", "horas_estudio": 5, "nota": 8, "usa_ia": "Sí"},
    {"nombre": "Juan", "edad": 28, "trabaja": "Sí", "horas_estudio": 2, "nota": 5, "usa_ia": "No"},
    {"nombre": "Sofía", "edad": 20, "trabaja": "No", "horas_estudio": 6, "nota": 9, "usa_ia": "Sí"},
    {"nombre": "Pedro", "edad": 24, "trabaja": "No", "horas_estudio": 3, "nota": 6, "usa_ia": "No"},
    {"nombre": "Lucía", "edad": 31, "trabaja": "Sí", "horas_estudio": 4, "nota": 7, "usa_ia": "Sí"},
    {"nombre": "Martín", "edad": 19, "trabaja": "No", "horas_estudio": 2, "nota": 4, "usa_ia": "No"},
    {"nombre": "Carla", "edad": 26, "trabaja": "Sí", "horas_estudio": 7, "nota": 9, "usa_ia": "Sí"},
    {"nombre": "Diego", "edad": 23, "trabaja": "No", "horas_estudio": 4, "nota": 7, "usa_ia": "Sí"},
]
```

## Actividad 1: convertir los datos en una tabla

1.   Elemento de la lista
2.   Elemento de la lista



Importá pandas y convertí la lista `estudiantes` en un DataFrame llamado `df`.

Luego mostrá la tabla.

```python
# Escribí tu código acá
import pandas as pd
df = pd.DataFrame(estudiantes)
print("Datos en formato lista: ")
print()
print(df)
```

## Actividad 2: calcular indicadores generales

Calculá y mostrá:

1. cantidad total de estudiantes;
2. promedio de edad;
3. promedio de nota;
4. promedio de horas de estudio.

```python
# Escribí tu código acá
total_estudiantes = len(estudiantes)
promedio_edad = df["edad"].mean()
promedio_nota = df["nota"].mean()
promedio_estudio = df["horas_estudio"].mean()

print(f"Cantidad total de estudiantes: {total_estudiantes}")
print(f"Promedio edades de los estudiantes: {promedio_edad}")
print(f"Promedio notas de los estudiantes: {promedio_nota}")
print(f"Promedio de las horas de estudio de los estudiantes: {promedio_estudio}")
```

## Actividad 3: aprobados y desaprobados

Consideramos aprobado a quien tiene nota igual o mayor a 6.

Creá una nueva columna llamada `estado`, con los valores:

- `Aprobado`
- `Desaprobado`

Luego contá cuántos estudiantes aprobaron y cuántos desaprobaron.

```python
# Escribí tu código acá
import numpy as np

resultados = []
for indice, fila in df.iterrows():
  if fila["nota"] >= 6:
    resultados.append("Aprobado")
  else:
    resultados.append("Desaprobado")

df["estado"] = resultados


print(df[["nombre", "nota", "estado"]])

conteo = df["estado"].value_counts()
print()
print("Cantidad aprobados y desaprobados: ", conteo)
```

## Actividad 4: trabajo y rendimiento

Separá los estudiantes en dos grupos:

- quienes trabajan;
- quienes no trabajan.

Luego calculá el promedio de nota de cada grupo.

```python
# Escribí tu código acá
quienes_trabajan = df[df["trabaja"] == "Sí"]
quienes_no_trabajan = df[df["trabaja"] == "No"]

print("Estudiantes que si trabajan")
print(quienes_trabajan[["nombre", "trabaja", "nota"]])

print("Estudiantes que no trabajan")
print(quienes_no_trabajan[["nombre", "trabaja", "nota"]])

print(f"Promedio de quienes trabajan: {quienes_trabajan['nota'].mean()}")
print(f"Promedio quienes no trabajan: {quienes_no_trabajan['nota'].mean()}")
```

## Actividad 5: uso de IA

Contá:

- cuántos estudiantes usan IA;
- cuántos estudiantes no usan IA.

```python
# Escribí tu código acá
usan = 0
no_usan = 0

for indice, fila in df.iterrows():
 if fila["usa_ia"] == "Sí":
  usan = usan + 1
 else:
  no_usan = no_usan + 1

print(f"Estudiantes que usan IA: {usan}")
print(f"Estudiantes que no usan IA: {no_usan}")
```

## Actividad 6: nota más alta y nota más baja

Mostrá:

- el nombre del estudiante con la nota más alta;
- el nombre del estudiante con la nota más baja.

```python
# Escribí tu código acá
import numpy as np

df_mayor_a_menor = df.sort_values(by="nota", ascending=False)
mas_alta = df_mayor_a_menor.head(1)


df_menor_a_mayor = df.sort_values(by="nota", ascending=True)
mas_baja = df_menor_a_mayor.head(1)

print(f"La nota mas alta es de: {mas_alta["nombre"].values[0]} (Nota:{mas_alta["nota"].values[0]})")
print(f"La nota mas baja es de: {mas_baja["nombre"].values[0]} (Nota:{mas_baja["nota"].values[0]})")
```

## Actividad 7: gráficos

Realizá al menos dos gráficos:

1. gráfico de barras con las notas de cada estudiante;
2. gráfico de dispersión con horas de estudio y nota final.

Podés usar `matplotlib`.

```python
# Escribí tu código acá
import matplotlib.pyplot as plt

nombres = df["nombre"]
notas = df["nota"]

plt.bar(nombres, notas)
plt.title("Notas de cada estudiante")
plt.xlabel("Estudiantes")
plt.ylabel("Notas")
plt.show()

horas = df["horas_estudio"]
notas_finales = df["nota"]

plt.scatter(horas, notas_finales)
plt.title("Relación entre Horas de Estudio y Nota Final")
plt.xlabel("Horas de Estudio")
plt.ylabel("Nota Final")
plt.show()
```
